```python
profits_df = data[['profits', 'rank']]
profits_df.plot.scatter('profits', 'rank')
```

```python
assets_df = data[['assets', 'rank']]
assets_df.plot.scatter('assets', 'rank')
# moderate
```

```python
return_df = data[['return_on_invested_capital_ROIC', 'rank']]
return_df.plot.scatter('return_on_invested_capital_ROIC', 'rank')
```

```python
revenues_df = data[['revenues', 'rank']]
revenues_df.plot.scatter('revenues', 'rank')
```

```python
CGR_df = data[['three_yr_CGR_revenues', 'rank']]
CGR_df.plot.scatter('three_yr_CGR_revenues', 'rank')
```

```python
# location based on country to make profit
plt.figure(figsize=(20,10))
data.groupby('state_country')['profits'].median().sort_values(ascending=False).plot.bar(color='b')
```

```python
# location based on geographic location to make profit
data.groupby('geographic_loc')['profits'].median().sort_values(ascending=False).plot.bar(color='b')
```

```python
# country to get max_ returns on investment
plt.figure(figsize=(20,10))
data.groupby('state_country')['return_on_invested_capital_ROIC'].median().sort_values(ascending=False).plot.bar(color='b')
```

```python
# country that has more assets ie popular in_ assets
plt.figure(figsize=(20,10))
data.groupby('state_country')['assets'].median().sort_values(ascending=False).plot.bar(color='b')
```

```python
plt.figure(figsize=(20,10))
data.groupby('state_country')['three_yr_CGR_revenues'].median().sort_values(ascending=False).plot.bar(color='b')
```
